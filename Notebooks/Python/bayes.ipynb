# Bayes

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# 🔹 Definir valores de θ (probabilidad de cara)
theta_vals = np.linspace(0, 1, 100)

# 🔹 Definir Prior (suponemos uniforme, es decir, todos los valores son igualmente probables)
prior = np.ones_like(theta_vals)  # Prior uniforme

# 🔹 Definir Verosimilitud (Likelihood) para 8 caras en 10 lanzamientos
n, k = 10, 8  # 10 lanzamientos, 8 caras
likelihood = stats.binom.pmf(k, n, theta_vals)  # Distribución binomial

# 🔹 Calcular Posterior (proporcional a Prior * Likelihood)
posterior = likelihood * prior
posterior /= np.trapz(posterior, theta_vals)  # Normalizar

# 🔹 Graficar los resultados
plt.figure(figsize=(10, 5))
plt.plot(theta_vals, prior, label="Prior", linestyle="dashed")
plt.plot(theta_vals, likelihood, label="Likelihood", linestyle="dotted")
plt.plot(theta_vals, posterior, label="Posterior", linewidth=2)
plt.xlabel("Probabilidad de Cara (θ)")
plt.ylabel("Densidad de Probabilidad")
plt.legend()
plt.title("Inferencia Bayesiana: Probabilidad de Cara tras 8 éxitos en 10 lanzamientos")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# 🔹 Definir valores de θ
theta_vals = np.linspace(0, 1, 100)

# 🔹 Calcular la verosimilitud para cada θ
n, k = 10, 8
likelihood = stats.binom.pmf(k, n, theta_vals)

# 🔹 Normalizar para escalar la verosimilitud a su valor máximo real (0.302)
likelihood_scaled = likelihood / np.max(likelihood) * 0.302

# 🔹 Graficar ambas escalas
plt.figure(figsize=(10, 5))
plt.plot(theta_vals, likelihood, label="Verosimilitud sin escalar", color="blue")
plt.plot(theta_vals, likelihood_scaled, label="Verosimilitud escalada (máximo = 0.302)", color="red", linestyle="dashed")
plt.xlabel("Probabilidad de Cara (θ)")
plt.ylabel("Densidad de Probabilidad")
plt.title("Comparación de Escalado en la Verosimilitud")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Parámetros de la normal
mu, sigma = 1.70, 0.05

# Valores de altura
x = np.linspace(1.5, 1.9, 100)

# Densidad de probabilidad
pdf = stats.norm(mu, sigma).pdf(x)

# Graficar
plt.figure(figsize=(8, 5))
plt.plot(x, pdf, label="Densidad de Probabilidad", color="blue")
plt.xlabel("Altura (m)")
plt.ylabel("Densidad de Probabilidad")
plt.title("Ejemplo de Densidad de Probabilidad")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# 🔹 Definir la media y desviación estándar de la altura
mu, sigma = 1.70, 0.05

# 🔹 Definir valores de altura
x_vals = np.linspace(1.50, 1.90, 100)

# 🔹 Calcular la densidad de probabilidad (PDF) para cada altura
pdf_vals = stats.norm(mu, sigma).pdf(x_vals)

# 🔹 Graficar la densidad de probabilidad
plt.figure(figsize=(8, 5))
plt.plot(x_vals, pdf_vals, label="Densidad de Probabilidad", color="blue")
plt.axvline(mu, color="red", linestyle="dashed", label="Media (1.70 m)")
plt.xlabel("Altura (m)")
plt.ylabel("Densidad de Probabilidad")
plt.title("Densidad de Probabilidad de la Altura")
plt.legend()
plt.show()
```
